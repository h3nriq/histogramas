# Aprenda Ciência de Dados: Histogramas 

Os histogramas gerados nesse notebook são baseados em dados aleatórios, portanto, a cada execução, os gráficos resultantes serão diferentes. Eles servem apenas como exemplos ilustrativos.

## Histograma de Distribuição Normal

```python
# Importa as bibliotecas necessárias
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Define a média e o desvio padrão para a distribuição normal
mean = 50
std_dev = 15

# Gera dados aleatórios com uma distribuição normal
normal_data = np.random.normal(mean, std_dev, 1000)

# Ajusta os dados para garantir que todos estejam dentro do intervalo de 0 a 100
normal_data = np.clip(normal_data, 0, 100)

# Configura o estilo do Seaborn
sns.set(style="whitegrid")

# Cria o histograma com Seaborn
ax = sns.histplot(normal_data, bins=20, kde=True, color="blue", edgecolor='black')

# Define os títulos e rótulos dos eixos
ax.set(title='Histograma', xlabel='Valor', ylabel='Densidade')

# Exibe o gráfico
plt.show()
```

## Histograma de Assimetria Negativa

```python
# Importa as bibliotecas necessárias
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Definindo parâmetros para a distribuição log-normal
mu = 3.5
sigma = 0.8

# Gerando dados com uma distribuição log-normal para criar assimetria negativa
negative_log_normal_data = np.random.lognormal(mu, sigma, 1000)

# Refletindo os dados em torno de um valor central (100) para criar a assimetria negativa
reflected_data = 100 - (negative_log_normal_data - np.min(negative_log_normal_data))

# Configurando o estilo do Seaborn
sns.set(style="whitegrid")

# Criando o histograma com Seaborn
ax = sns.histplot(reflected_data, bins=30, kde=True, color="blue", edgecolor='black')

# Define os títulos e rótulos dos eixos
ax.set(title='Histograma com Assimetria Negativa', xlabel='Valor', ylabel='Densidade')

# Exibe o gráfico
plt.show()
```

## Moda, Média e Mediana

```python
# Importa as bibliotecas necessárias
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd

# Parâmetros para a distribuição log-normal
mu = 3.5
sigma = 0.8

# Gerando dados com uma distribuição log-normal
log_normal_data = np.random.lognormal(mu, sigma, 1000)

# Calculando a média e a mediana dos dados
mean = np.mean(log_normal_data)
median = np.median(log_normal_data)

# Categorizando os dados em intervalos
bins = np.linspace(min(log_normal_data), max(log_normal_data), 20)
intervals = pd.cut(log_normal_data, bins)
mode_interval = intervals.value_counts().idxmax()

# Configurando o estilo do Seaborn
sns.set(style="whitegrid")

# Criando o histograma com Seaborn
ax = sns.histplot(log_normal_data, bins=20, kde=True, color="blue", edgecolor='black')

# Adicionando linhas verticais para a média, mediana e moda
plt.axvline(mean, color='r', linestyle='--', label=f'Média: {mean:.2f}')
plt.axvline(median, color='g', linestyle='-', label=f'Mediana: {median:.2f}')
plt.axvline(mode_interval.left, color='purple', linestyle='-', label=f'Moda (intervalo): {mode_interval}')

# Adicionando a legenda ao gráfico
plt.legend(fontsize=8, loc='upper right')

# Definindo os títulos e rótulos dos eixos
ax.set(title='Histograma Assimétrico Positivo - Local da Moda, Média e Mediana', xlabel='Valor', ylabel='Densidade')

# Exibindo o gráfico
plt.show()
```

## Histograma Bimodal

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Parâmetros das duas distribuições normais
mu1, sigma1 = 20, 4
mu2, sigma2 = 40, 4

# Gerando as duas distribuições normais
normal_data1 = np.random.normal(mu1, sigma1, 1000)
normal_data2 = np.random.normal(mu2, sigma2, 1000)

# Combinando as duas distribuições para formar uma distribuição bimodal
bimodal_data = np.concatenate([normal_data1, normal_data2])

# Criando o histograma
plt.figure(figsize=(10, 6))
ax = sns.histplot(bimodal_data, bins=30, kde=False, color="blue", edgecolor='black')

# Marcando aproximadamente as regiões das modas
plt.axvline(mu1, color='r', linestyle='--', label=f'Moda Aproximada 1: {mu1}')
plt.axvline(mu2, color='g', linestyle='--', label=f'Moda Aproximada 2: {mu2}')

# Configurando o gráfico
plt.title('Histograma - Distribuição Bimodal')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.legend()

# Exibindo o gráfico
plt.show()
```
